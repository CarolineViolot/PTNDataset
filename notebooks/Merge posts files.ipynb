```python
import json
import os
import sys
from tqdm.notebook import tqdm
if 'notebooks' in os.listdir('../'):
    os.chdir('../')
```

```python
POST_FOLDER = 'test_data/posts'
OUTPUT_POST = 'test_data/posts.json'
channel_files = [f for f in os.listdir(POST_FOLDER) if ".json" in f]

pbar = tqdm(channel_files, desc='Reading channels', unit='file')
posts = dict()
for channel_filename in pbar:
    pbar.set_postfix_str(channel_filename)
    channel_url = channel_filename.replace('.json', '').replace('telegram_messages-', '')
    with open(os.path.join(POST_FOLDER, channel_filename), 'r') as f:
        data = json.load(f, parse_int=str)
    posts[channel_url] = data
```

```python
with open(OUTPUT_POST, 'w') as f:
    json.dump(posts, f)
```
